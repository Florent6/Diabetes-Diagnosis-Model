```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv("../dataset/diabetes_cleaned.csv")

# Set plot style
sns.set(style="darkgrid")
plt.rcParams['figure.figsize'] = (10, 6)

#univariate distributions
numeric_cols = ['age', 'bmi', 'hba1c_level', 'blood_glucose_level']

for col in numeric_cols:
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col.capitalize()}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

#diabetes count
sns.countplot(data=df, x='diabetes')
plt.title("Diabetes Diagnosis Count")
plt.xticks([0, 1], ['No Diabetes', 'Diabetes'])
plt.ylabel("Patients")
plt.show()

#categorical vs target (diabates)
sns.countplot(data=df, x='smoking_history', hue='diabetes')
plt.title("Diabetes by Smoking History")
plt.xticks(rotation=45)
plt.ylabel("Patients")
plt.show()

sns.boxplot(data=df, x='diabetes', y='bmi')
plt.title("BMI by Diabetes")
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()

#correlation heatmap
corr = df[numeric_cols + ['diabetes']].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
